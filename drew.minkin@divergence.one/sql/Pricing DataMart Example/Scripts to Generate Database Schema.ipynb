```sql
/****** Object:  UserDefinedFunction [dbo].[StripToNumeric]    Script Date: 5/17/2022 11:50:23 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE FUNCTION [dbo].[StripToNumeric] (@INPUT varchar(100))
RETURNS int
WITH EXECUTE AS CALLER
AS
BEGIN
	DECLARE @StripToNumeric int
	DECLARE @Iteration int
	DECLARE @Loop int
	DECLARE @check char(1)

     SET @Iteration = 0
	WHILE @Iteration < LEN(@INPUT)
	BEGIN
			SET @Iteration = @Iteration + 1
	set @check = substring(@input,@Iteration,1)
	if charindex(@check,'0123456789') = 0
		begin
		set @INPUT = REPLACE(@INPUT,@check,'')
		END
	END
	SET @StripToNumeric = @INPUT
     RETURN(@StripToNumeric)
END
GO
```

# [dbo].[sp_logAndRethrowError]

```sql
/****** Object:  StoredProcedure [dbo].[sp_logAndRethrowError]    Script Date: 5/17/2022 11:50:23 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER OFF
GO
-- Create the stored procedure to generate an error using 
-- RAISERROR. The original error is logged into logTable.

create procedure [dbo].[sp_logAndRethrowError] 
as 
begin 
    -- return if there is no error information to retrieve.
    if error_number() is null
        return;

    declare 
        @ErrorMessage    nvarchar(4000),
        @ErrorNumber     int,
        @ErrorSeverity   int,
        @ErrorState      int,
        @ErrorLine       int,
        @ErrorProcedure  nvarchar(200);


    -- Assign variables to error-handling functions that 
    select 
        @ErrorNumber = ERROR_NUMBER(),
        @ErrorSeverity = ERROR_SEVERITY(),
        @ErrorState = ERROR_STATE(),
        @ErrorLine = ERROR_LINE(),
        @ErrorProcedure = isnull(ERROR_PROCEDURE(), '-');


    -- Building the message string with original error information.
    set @ErrorMessage = N'Error Number ' + cast(@ErrorNumber AS nvarchar(50))+ ', ' +
                         'Error Line '   + cast(@ErrorLine AS nvarchar(50)) + ', ' +
                         'Error Message : ''' + error_message() + ''''
                       
                                 
    -- Logging the error information into logTable if it isn't logged yet
    if @ErrorState < 127 
        exec sp_logWrite @logLevel = 'FATAL', 
                         @processName = @ErrorProcedure, 
                         @logText = @ErrorMessage      


    -- Raising an error
    raiserror(
        @ErrorMessage, 
        @ErrorSeverity, 
        127 -- indicates that error is thrown mannualy 
    );
end
GO
```

# [dbo].[sp_logBatch]

```sql
/****** Object:  StoredProcedure [dbo].[sp_logBatch]    Script Date: 5/17/2022 11:50:23 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[sp_logBatch] (@action nvarchar(8), @activity nvarchar(50) = NULL) as

set nocount on
declare @recCount integer
-- Start
if @action = N'Start' begin
   insert logBatch (startTime, activity)
            values (getdate(), coalesce(@activity, N'ETL'))
end
else begin
  -- Stop
  if @action = N'Stop' begin
     update logBatch
        set endTime = getdate()
      where batchID = (select max(batchID) from logBatch where activity = coalesce(@activity, activity))
        
  end
  else begin
      -- error
      RAISERROR (N'Invalid Action parameter <%s> for sp_logBatch',
                  16, 1, @action)
      return(-2)
  end
end
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[sp_logBeginSection]    Script Date: 5/17/2022 11:50:23 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER OFF
GO
create procedure [dbo].[sp_logBeginSection] (@section nvarchar(1000)) as


set nocount on
SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED
declare @currentSection nvarchar(1000)
declare @separator nchar(1)
set @separator = N'.'
select @currentSection = section
  from logSession
 where spID = @@SPID
-- record not found hence initialize it
if @@ROWCOUNT = 0
  begin
    exec sp_logInitialize
    set @currentSection = NULL
  end
if @currentSection is NULL or len(@currentSection) = 0
 begin
   UPDATE logSession
     set section = @section
   where spID = @@SPID
 end
else
 begin
   UPDATE logSession
     set section = @currentSection + @separator + @section
   where spID = @@SPID
 end
GO
```

# [dbo].[sp_logEndSection]

```sql
/****** Object:  StoredProcedure [dbo].[sp_logEndSection]    Script Date: 5/17/2022 11:50:23 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER OFF
GO
create procedure [dbo].[sp_logEndSection] (@section nvarchar(1000) = NULL) as

set nocount on
SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED
declare @currentSection nvarchar(1000)
declare @pos integer
-- if input is NULL then clear all sections
if @section is NULL
 begin
  UPDATE logSession
    set section = NULL
  where spID = @@SPID
  return
 end
select @currentSection = section 
 from logSession
where spID = @@SPID
set @pos = charindex(@section, @currentSection)
-- section not found, no action required
if @pos = 0 or @pos is NULL
  return
-- also remove 1 char separator '.'. If first one, then clear it.
if @pos = 1
  begin
   set @currentSection = NULL
  end
else
  begin
   set @currentSection = substring(@currentSection, 1, @pos - 2)
  end
UPDATE logSession
  set section = @currentSection
where spID = @@SPID
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[sp_logSeverityProcessing]    Script Date: 5/17/2022 11:50:23 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER OFF
GO
create procedure [dbo].[sp_logSeverityProcessing] (@errorFlag integer,
                                            @processName nvarchar(50),
                                            @logText nvarchar(2048) = NULL) as

set nocount on
SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED
declare @abortProcess nchar(1)
declare @logID integer
declare @subject nvarchar(200)
-- validate severity error and abort if failure (less than zero)
if @errorFlag = 0
  return
select @abortProcess=abortProcess
 from logErrorSeverity
where errorFlag = @errorFlag and abortProcess = N'Y'
if @@ROWCOUNT = 0 begin
   return(0)
end
-- check if abort process required
select @logID = max(logID) from logTable where processName = @processName
if @logID is Null
  set @logID = -1
set @subject = N'***ZPRO Fatal error logged for processName = ' +
               @processName + N', logID = ' + cast(@logID as nvarchar)
exec sp_logEmail @subject = @subject,
                 @message = @logText
RAISERROR (N'Serious error found for processName = %s, error severity = %d in sp_logSeverityProcessing',
           16, 1, @processName, @errorFlag)
return(@errorFlag)
GO
```

# [dbo].[sp_logWrite]

```sql
/****** Object:  StoredProcedure [dbo].[sp_logWrite]    Script Date: 5/17/2022 11:50:23 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER OFF
GO
create procedure [dbo].[sp_logWrite] (@logLevel nvarchar(8) = 'INFO',
                               @processName nvarchar(50) = NULL,
                               @logText nvarchar(2048) = NULL,
                               @action nvarchar(8) = 'LogInfo',
                               @tblName nvarchar(50) = NULL,
                               @colName nvarchar(50) = NULL,
                               @tableRowCount integer = 0,
                               @insertCount integer = 0,
                               @updateCount integer = 0,
                               @rejectCount integer = 0,
                               @deleteCount integer = 0,
                               @errorFlag integer = 0,
                               @number1 float = NULL,
                               @number2 float = NULL,
                               @number3 float = NULL,
                               @reportLevel tinyint = 0
                               ) as


set nocount on
SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED
declare @errorSeverityID integer
declare @batchID integer
declare @logID integer
declare @rowCount integer
declare @currTime datetime
declare @defaultLogLevel nvarchar(25)
declare @defaultLevelNo int
declare @currentLevelNo int
declare @allLogLevels nvarchar(40)
declare @section nvarchar(1000)
-- check @action
if @action not in (N'Start', N'Stop', N'LogInfo', N'sqlDebug') begin
   RAISERROR (N'action <%s> is invalid, correct code',
              16, 1, @action)
end
-- check logLevel, if not greater or equal to current default level, then return with no logging.
-- some fancy string manipulation to set the order
set @allLogLevels = N'OFFFATALERRORWARNINFODEBUGALL'
set @currentLevelNo = charindex(@logLevel, @allLogLevels)
if @currentLevelNo = 0 begin
   RAISERROR (N'sp_logWrite.logLevel <%s> is invalid, pass the correct level parameter',
              16, 1, @logLevel)
   return(1)
end
exec sp_logGetLevel @defaultLogLevel OUTPUT
if @defaultLogLevel = 'OFF' or @logLevel = 'OFF' begin
   return(0)
end
set @defaultLevelNo = charindex(@defaultLogLevel, @allLogLevels)
if @defaultLevelNo = 0 begin
   RAISERROR (N'logSession.logLevel <%s> is not set-up, set it up',
              16, 1, @defaultLogLevel)
   return(1)
end
-- check if logging level is > than default level
if @currentLevelNo > @defaultLevelNo begin
   return(0)
end
--
-- Stop Log - Must have Start record
if @action = N'Stop' begin
-- get record to be updated
  select @logID = max(logID)
  from logTable
  where processName = @processName and action = N'Start'
  if @logID is not null begin
  -- now update record
    UPDATE logTable
      set endTime = getdate(),
          action = @action,
          tblName = @tblName,
          colName = @colName,
          logText = @logText,
          tableRowCount = @tableRowCount,
          insertCount = @insertCount,
          updateCount = @updateCount,
          rejectCount = @rejectCount,
          deleteCount = @deleteCount,
          errorFlag = @errorFlag,
          number1 = @number1,
          number2 = @number2,
          number3 = @number3,
          reportLevel = @reportLevel
    where logID = @logID
    if @@ROWCOUNT = 0 begin
      RAISERROR (N'processName <%s> update failed in sp_logWrite',
                16, 1, @processName)
      return(-1)
    end
    -- check if abort required.
    exec sp_logSeverityProcessing @errorFlag = @errorFlag,
                                  @processName = @processName,
                                  @logText = @logText
    return 0
  end
  else begin
  RAISERROR (N'Expecting Log record to update for processName <%s>',
              16, 1, @processName)
  end
end
--
-- Start or LogInfo process
-- get batchID
select @batchID = max(batchID) from logBatch
-- get section
exec sp_logGetSection @section=@section OUTPUT
-- set time
if @action = N'Start' begin
   set @currTime = NULL
end
else begin
   set @currTime = getdate()
end
--write record to the table
insert INTO logTable
             (batchID, logLevel, processName, action, endTime,
              logText, section, tblName, colname, errorFlag,
              tableRowCount, insertCount, updateCount, rejectCount, deleteCount,
              number1, number2, number3, reportLevel, spID)
      VALUES (@batchID, @logLevel, @processName, @action, @currTime,
              @logText, @section, @tblName, @colName, @errorFlag,
              @tableRowCount, @insertCount, @updateCount, @rejectCount, @deleteCount,
              @number1, @number2, @number3, @reportLevel, @@SPID)
if @@ROWCOUNT = 0 begin
   RAISERROR (N'processName <%s> insert failed',
              16, 1, @processName)
   return(-1)
end
-- check if abort required
exec sp_logSeverityProcessing @errorFlag = @errorFlag,
                              @processName = @processName,
                              @logText = @logText
return 0
GO
```

# [dbo].[spr_dimProcessingCustomers]

```sql
/****** Object:  StoredProcedure [dbo].[spr_dimProcessingCustomers]    Script Date: 5/17/2022 11:50:23 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[spr_dimProcessingCustomers] as 
begin

    /*
    exec spr_dimProcessingCustomers

    select * from customers
    select * from DimCustomer
    select * from logTable where processName = 'spr_dimProcessingCustomers'
    truncate table logTable
    */

/*
Goals
1.	Insert customers from pricingstaging
2.	Get undefined customers from quotes
3.	Get undefined customers from orders
4.	Apply customer velocity
5.	Apply customer chain district key
6.	Apply customer chain key
7.	Update unknown dimension attributes
*/
    begin try
        declare @processName varchar(100)
        declare @tableName varchar(50)
        declare @stgTableRowCount int
        declare @dmRowCountBeforeProcessing int
        declare @dmFinalTableRowCount int
        declare @logID int
        declare @insertCount int
        declare @updateCount int
        declare @currentDate datetime
		declare @mediantop int
		declare @medianbottom int
		declare @uppertop int
		declare @upperbottom int
		declare @lowertop int
		declare @lowerbottom int
		declare @salesmedian numeric(38,16)
		declare @saleslower numeric(38,16)		
		declare @salesupper numeric(38,16)

        -- initialize parameters
        set @processName = 'spr_dimProcessingCustomers'
        set @tableName = 'DimCustomer'
        set @stgTableRowCount = (select count(Distributor_id) from pricingstaging..Customer)
        set @dmRowCountBeforeProcessing = (select count(customerKey) from DimCustomer)
        set @currentDate = getDate()

        /*
        Initialize the logs for this procedure
        EVERY STORED PROCEDURES SHOULD INCLUDE AT LEAST A START AND STOP LOG ENTRY
        */
        exec sp_logBeginSection @processName
        exec sp_logWrite @action = 'START', 
		        @logLevel = 'INFO', 
		        @processName = @processName, 
		        @tblName = @tableName,
		        @logText = @processName,
		        @number1 = @stgTableRowCount,
		        @number2 = @dmRowCountBeforeProcessing
        -- get the logid from the logTable for this process
        set @logID = (select max(logID) from logTable where processName = @processName)


        /*--------------------------------------------------------------------------------------------*/
        /*------1.	Insert customers from pricingstaging-----------------------------------------------------*/
        /*--------------------------------------------------------------------------------------------*/

        /*
        First insert new records into the pricingstaging table where dimension records do not exist.
        This should be implemented for every dimenion being processed whether it is a type 1 or type 2 dimension.
        */
        exec sp_logBeginSection 'New Dimension Records'
        exec sp_logWrite @action = 'START', 
		        @logLevel = 'INFO', 
		        @processName = @processName, 
		        @tblName = @tableName,
		        @logText = @processName
        -- get the logid from the logTable for this process
        set @logID = (select max(logID) from logTable where processName = @processName)
		truncate table DimCustomer 
        insert into DimCustomer (
													CustomerNumber,--1
													CustName,--2
													CustChainCode, --3
													CustChainDescr, --4
													CustDistCode, --5
                                                    CustDistDesc, --6
													CustRegCode, --7
													CustRegDesc, --8
													SalesOfficeCode, --9
													SalesOfficeDesc, --10
													ParentCustomerNumber, --11
													BusinessUnitId, --12
													DomSalesOffice, --13
                                                    CustVelocity, --14
													RowIsCurrent, --15
													RowStartDate, --16
													RowEndDate, --17
													LogID,
													Source)--18
			select stg.Distributor_id,--1
					stg.Name,--2
					stg.Chain_id,--3
					stg.Chain_name,--4
					stg.District_id,--5
					stg.District_name,--6
					stg.Region_id,--7
					stg.Region_desc,--8
					stg.Sales_office_id,--9
					stg.Sales_office_desc,--10
					stg.Comp_sheet,--11
					stg.business_unit_id,--12
					-1, --stg.DomSalesOfficeBucket,--13
					-1, --stg.CustVelocity--14
					1,--15
					getdate(),--@currentDate,--16
					'12-31-2099',--17
					@logID,
					'Extract' --18
			from pricingstaging..Customer stg

        -- get the row count to include in the logging
        set @insertCount = @@rowcount

        /*
        Log the number of new records inserted into the dimension table
        */
        exec sp_logWrite @action = 'STOP',
		        @logLevel = 'INFO', 
		        @processName = @processName, 
		        @tblName = @tableName,
		        @logText = @processName,
		        @insertCount = @insertCount
        exec sp_logEndSection 'New Dimension Records'
        		

        /*--------------------------------------------------------------------------------------------*/
        /*------2.	Get undefined customers from quotes-----------------------------------------------*/
        /*--------------------------------------------------------------------------------------------*/
        exec sp_logBeginSection 'Get undefined customers from quotes'
        exec sp_logWrite @action = 'START', 
		        @logLevel = 'INFO', 
		        @processName = @processName,
		        @tblName = @tableName, 
		        @logText = @processName

		select  
		CUSTOMER_NUM	distributor_id,
		AGENT			sales_office_id,
		REGION			region_id,
		MKT_GROUP_ID	Business_unit_id,
		count(*) qty
		into #quotecustomer
		from pricingstaging.dbo.quotes
		group by 
		AGENT,
		REGION,
		MKT_GROUP_ID ,
		CUSTOMER_NUM
	
		delete from #quotecustomer 
		from #quotecustomer q
		join dimcustomer c
		on  c.customernumber	=q.distributor_id	
		AND	c.SalesOfficeCode	=q.sales_office_id
		AND	c.CustRegCode			=q.region_id	
		AND c.BusinessUnitId	=q.Business_unit_id

		insert into 
		DimCustomer
		(
		CustomerNumber,				
		SalesOfficeCode,
		CustRegCode,						
		BusinessUnitId,
		Source
		)
		SELECT
		distributor_id,				
		sales_office_id,
		region_id,						
		Business_unit_id,
		'Quotes'
		FROM
		#quotecustomer

        set @insertCount = @@rowcount

		UPDATE DimCustomerQuotes
		SET CustName = DimCustomerExtract.CustName
		FROM (SELECT * from DimCustomer where Source = 'Quotes') DimCustomerQuotes
		JOIN (SELECT * from DimCustomer where Source = 'Extract') DimCustomerExtract
		ON DimCustomerQuotes.CustomerNumber	= DimCustomerExtract.CustomerNumber				

		UPDATE DimCustomerQuotes
		SET CustRegDesc = DimCustomerExtract.CustRegDesc
		FROM (SELECT * from DimCustomer where Source = 'Quotes') DimCustomerQuotes
		JOIN (SELECT * from DimCustomer where Source = 'Extract') DimCustomerExtract
		ON DimCustomerQuotes.CustRegCode	= DimCustomerExtract.CustRegCode				

		UPDATE DimCustomerQuotes
		SET SalesOfficeDesc = DimCustomerExtract.SalesOfficeDesc
		FROM (SELECT * from DimCustomer where Source = 'Quotes') DimCustomerQuotes
		JOIN (SELECT * from DimCustomer where Source = 'Extract') DimCustomerExtract
		ON DimCustomerQuotes.SalesOfficeCode	= DimCustomerExtract.SalesOfficeCode				

        -- log the update
        exec sp_logWrite @action = 'STOP', 
		        @logLevel = 'INFO', 
		        @processName = @processName, 
		        @tblName = @tableName,
		        @logText = @processName,
		        @insertCount = @insertCount
        exec sp_logEndSection 'Get undefined customers from quotes'

        /*--------------------------------------------------------------------------------------------*/
        /*------3.	Get undefined customers from orders-----------------------------------------------*/
        /*--------------------------------------------------------------------------------------------*/
        exec sp_logBeginSection 'Get undefined customers from orders'
        exec sp_logWrite @action = 'START', 
		        @logLevel = 'INFO', 
		        @processName = @processName,
		        @tblName = @tableName, 
		        @logText = @processName

		select  
		distributor_id,
		convert(varchar(50),agent_office_id) sales_office_id,
		region_id,
		Business_unit_id,
		sum(ext_price) salesdollars
		into #quoteorder
		from pricingstaging.dbo.orders
		group by 
		region_id,
		agent_office_id,
		Business_unit_id,
		distributor_id

		delete from #quoteorder 
		from #quoteorder q
		join dimcustomer c
		on  c.customernumber	=q.distributor_id	
		AND	c.SalesOfficeCode	=q.sales_office_id
		AND	c.CustRegCode			=q.region_id	
		AND c.BusinessUnitId	=q.Business_unit_id

		insert into 
		DimCustomer
		(
		CustomerNumber,				
		SalesOfficeCode,
		CustRegCode,						
		BusinessUnitId,
		Source
		)
		SELECT
		distributor_id,				
		sales_office_id,
		region_id,						
		Business_unit_id,
		'Orders'
		FROM
		#quoteorder

        set @insertCount = @@rowcount

		UPDATE DimCustomerOrders
		SET CustName = DimCustomerExtract.CustName
		FROM (SELECT * from DimCustomer where Source = 'Orders') DimCustomerOrders
		JOIN (SELECT * from DimCustomer where Source = 'Extract') DimCustomerExtract
		ON DimCustomerOrders.CustomerNumber	= DimCustomerExtract.CustomerNumber				

		UPDATE DimCustomerOrders
		SET CustRegDesc = DimCustomerExtract.CustRegDesc
		FROM (SELECT * from DimCustomer where Source = 'Orders') DimCustomerOrders
		JOIN (SELECT * from DimCustomer where Source = 'Extract') DimCustomerExtract
		ON DimCustomerOrders.CustRegCode	= DimCustomerExtract.CustRegCode				

		UPDATE DimCustomerOrders
		SET SalesOfficeDesc = DimCustomerExtract.SalesOfficeDesc
		FROM (SELECT * from DimCustomer where Source = 'Orders') DimCustomerOrders
		JOIN (SELECT * from DimCustomer where Source = 'Extract') DimCustomerExtract
		ON DimCustomerOrders.SalesOfficeCode	= DimCustomerExtract.SalesOfficeCode	

        -- log the update
        exec sp_logWrite @action = 'STOP', 
		        @logLevel = 'INFO', 
		        @processName = @processName, 
		        @tblName = @tableName,
		        @logText = @processName,
		        @insertCount = @insertCount
        exec sp_logEndSection 'Get undefined customers from orders'


        /*--------------------------------------------------------------------------------------------*/
        /*----------4. Apply customer velocity--------------------------------------------------------*/
        /*--------------------------------------------------------------------------------------------*/

        exec sp_logBeginSection 'Apply customer velocity'
        exec sp_logWrite @action = 'START', 
		        @logLevel = 'INFO', 
		        @processName = @processName,
		        @tblName = @tableName, 
		        @logText = @processName
				
				set @currentDate = dateadd(dd,-365,(select max(invoice_date) from pricingstaging.dbo.orders))

				select 
				row_number() OVER(order by sum(ext_price)) rank, 
				distributor_id customerkey, 
				sum(ext_price) sales
				into #customervelocity
				from pricingstaging.dbo.orders
				where invoice_date < @currentDate
				and line_type = 'I'
				group by distributor_id

			
				set @medianbottom 
				= 	case	when @@rowcount % 2 = 1 then ((@@rowcount-1)/2 + 1)
					else 	(@@rowcount)/2 - 1 end
				
				set @mediantop 
				= 	case	when @@rowcount % 2 = 1 then ((@@rowcount-1)/2 + 1)
					else 	((@@rowcount)/2)  end

				if @mediantop = @medianbottom
					begin
							set @salesmedian = (select sales from #customervelocity where rank = @mediantop)
							set @salesupper = 
							case 
								when	@mediantop % 2 = 0
								then	(select sales from #customervelocity where rank = @mediantop/2)	
								else	(
										select sum(sales)/2 
										from #customervelocity 
										where rank in (((@mediantop - 1)/2),(((@mediantop - 1)/2) + 1))
										)
								end		
							set @saleslower = 
							case 
								when	@mediantop % 2 = 0
								then	(select sales from #customervelocity where rank = (3*@mediantop)/2)
								else	(
										select sum(sales)/2 
										from #customervelocity 
										where rank in (((3*(@mediantop - 1))/2),(((3*(@mediantop - 1))/2) + 1))
										)
								end		
					end
				else
							set @salesmedian = (select sum(sales)/2 from #customervelocity where rank in (@mediantop, @medianbottom))
							set @salesupper = 
							case 
								when	@mediantop % 2 = 0
								then	(select sales from #customervelocity where rank = (@mediantop+1)/2)
								else	(
										select sum(sales)/2 
										from #customervelocity 
										where rank in (((@mediantop)/2),(((@mediantop)/2) + 1))
										)
								end										
							set @saleslower =
 							case 
								when	@mediantop % 2 = 0
								then	(select sales from #customervelocity where rank = (3*(@mediantop+1))/2)
								else	(
										select sum(sales)/2 
										from #customervelocity 
										where rank in (((3*(@mediantop))/2),(((3*(@mediantop))/2) + 1))
										)
								end

			update dimcustomer
			set CustVelocity
			= CASE 
			WHEN	v.sales >= @salesupper							THEN	'High Discount' 
			WHEN	v.sales BETWEEN @salesmedian AND @salesupper	THEN	'Medium High Discount'
			WHEN	v.sales BETWEEN @saleslower AND @salesmedian	THEN	'Medium Low Discount'
			WHEN	v.sales <= @saleslower							THEN	'Low Discount'
			ELSE 'Unknown'
			END ,
			velocitykey = CASE 
			WHEN	v.sales >= @salesupper							THEN	1 
			WHEN	v.sales BETWEEN @salesmedian AND @salesupper	THEN	2
			WHEN	v.sales BETWEEN @saleslower AND @salesmedian	THEN	3
			WHEN	v.sales <= @saleslower							THEN	4
			ELSE 5
			END 
			from dimcustomer p
			left join #customervelocity v
			on p.customernumber =v.customerkey

		set @updateCount = @@rowcount

        exec sp_logWrite @action = 'STOP', 
		        @logLevel = 'INFO', 
		        @processName = @processName, 
		        @tblName = @tableName,
		        @logText = @processName,
		        @updateCount = @updateCount
        exec sp_logEndSection 'Apply customer velocity'


        /*--------------------------------------------------------------------------------------------*/
        /*-------5. Apply customer chain district key-------------------------------------------------*/
        /*--------------------------------------------------------------------------------------------*/

        exec sp_logBeginSection 'Apply customer chain district key'
        exec sp_logWrite @action = 'START', 
		        @logLevel = 'INFO', 
		        @processName = @processName,
		        @tblName = @tableName, 
		        @logText = @processName

		select CustDistCode, 0 as  chaindistrictkey 
		into #districts
		from dimcustomer 
		where isnumeric(CustDistCode) = 1 and CustDistCode is not null
		
		update #districts set chaindistrictkey = convert(int, CustDistCode) 

		update dimcustomer set chaindistrictkey = d.chaindistrictkey 
		from dimcustomer c
		join #districts  d
		on c.CustDistCode = d.CustDistCode 

		set @updateCount = @@rowcount
		set @uppertop = (select max(chaindistrictkey) from dimcustomer)


		select @uppertop + row_number() OVER (ORDER BY CustDistCode ) chaindistrictkey,  CustDistCode, count(*) qty
		into #alphadistkey
		from dimcustomer where chaindistrictkey is null and CustDistCode is not null
		group by CustDistCode

		set @upperbottom = @@rowcount


		update dimcustomer set chaindistrictkey = a.chaindistrictkey
		from dimcustomer c
		join #alphadistkey a on
		a.CustDistCode= c.CustDistCode
		
		set @updateCount = @updateCount  + @@rowcount

		update dimcustomer set chaindistrictkey = @upperbottom + @uppertop + 1
		where chaindistrictkey is null

		set @updateCount = @updateCount  + @@rowcount

		update dimcustomer set custdistcode = 
		CASE 
		when len(custdistcode) = 2 and isnumeric(custdistcode) = 1 then '00'+ custdistcode
		when len(custdistcode) = 3 and isnumeric(custdistcode) = 1 then '0'+ custdistcode
		END

        exec sp_logWrite @action = 'STOP', 
		        @logLevel = 'INFO', 
		        @processName = @processName, 
		        @tblName = @tableName,
		        @logText = @processName,
		        @updateCount = @updateCount
        exec sp_logEndSection 'Apply customer chain district key'


        /*--------------------------------------------------------------------------------------------*/
        /*-------6. Apply customer chain key----------------------------------------------------------*/
        /*--------------------------------------------------------------------------------------------*/

        exec sp_logBeginSection 'Apply customer chain key'
        exec sp_logWrite @action = 'START', 
		        @logLevel = 'INFO', 
		        @processName = @processName,
		        @tblName = @tableName, 
		        @logText = @processName


		select CustChainCode, 0 as  chainkey 
		into #chains
		from dimcustomer 
		where isnumeric(CustChainCode) = 1 and CustChainCode is not null
		
		update #chains set chainkey = convert(int, CustChainCode) 

		update dimcustomer set chainkey = d.chainkey 
		from dimcustomer c
		join #chains  d
		on c.CustChainCode = d.CustChainCode 

		set @updateCount = @@rowcount
		set @uppertop = (select max(chainkey) from dimcustomer)

		select @uppertop + row_number() OVER (ORDER BY CustChainCode ) chainkey,  CustChainCode, count(*) qty
		into #alphachainkey
		from dimcustomer where chainkey is null and CustChainCode is not null
		group by CustChainCode

		set @upperbottom = @@rowcount

		update dimcustomer set chainkey = a.chainkey
		from dimcustomer c
		join #alphachainkey a on
		a.CustChainCode= c.CustChainCode
		
		set @updateCount = @updateCount  + @@rowcount

		update dimcustomer set chainkey = @upperbottom + @uppertop + 1
		where chainkey is null

		set @updateCount = @updateCount  + @@rowcount

        exec sp_logWrite @action = 'STOP', 
		        @logLevel = 'INFO', 
		        @processName = @processName, 
		        @tblName = @tableName,
		        @logText = @processName,
		        @updateCount = @updateCount
        exec sp_logEndSection 'Apply customer chain key'

        /*--------------------------------------------------------------------------------------------*/
        /*-------7.	Update unknown dimension attributes-----------------------------------------------*/
        /*--------------------------------------------------------------------------------------------*/

        exec sp_logBeginSection 'Update unknown dimension attributes'
        exec sp_logWrite @action = 'START', 
		        @logLevel = 'INFO', 
		        @processName = @processName,
		        @tblName = @tableName, 
		        @logText = @processName
		
		UPDATE DimCustomer
		SET
		CustomerNumber  				= isnull(CustomerNumber  	,'zUnknown - CustomerNumber'),
		CustName        				= isnull(CustName        	,'zUnknown - CustName'),
		CustChainCode   				= isnull(CustChainCode   	,'zUnknown - CustChainCode'),
		CustChainDescr  				= isnull(CustChainDescr  	,'zUnknown - CustChainDesc'),
		CustDistCode    				= isnull(CustDistCode    	,'zUnknown - CustDistCode'),
		CustDistDesc    				= isnull(CustDistDesc    	,'zUnknown - CustDistDesc'),
		CustRegCode     				= isnull(CustRegCode     	,'zUnknown - CustRegCode'),
		CustRegDesc     				= isnull(CustRegDesc     	,'zUnknown - CustRegDesc'),
		SalesOfficeCode 				= isnull(SalesOfficeCode 	,'zUnknown - SalesOfficeCode'),
		SalesOfficeDesc 				= isnull(SalesOfficeDesc 	,'zUnknown - SalesOfficeDesc'),
		ParentCustomerNumber			= isnull(ParentCustomerNumber,'zUnknown -ParentCustomer'),
		BusinessUnitId  				= isnull(BusinessUnitId  	,'zUnknown - BusinessUnitId'),
		DomSalesOfficeCode  			= isnull(DomSalesOfficeCode  ,'zUnknown -DomSalesOfficeCode'),
		RowIsCurrent    				= 1,
		RowStartDate    				= getdate(),
		RowEndDate      				= '12-31-2099'


        exec sp_logWrite @action = 'STOP', 
		        @logLevel = 'INFO', 
		        @processName = @processName, 
		        @tblName = @tableName,
		        @logText = @processName,
		        @updateCount = @updateCount
        exec sp_logEndSection 'Update unknown dimension attributes'

        /*
        Finalize the logs for this procedure
        */
        set @dmFinalTableRowCount = (select count(customerKey) from DimCustomer)
        exec sp_logWrite @action = 'STOP', 
		        @logLevel = 'INFO', 
		        @processName = @processName, 
		        @logText = 'Successful Run', 
		        @tblName = @tableName,
		        @tableRowCount = @dmFinalTableRowCount,
		        @number1 = @stgTableRowCount,
		        @number2 = @dmRowCountBeforeProcessing
        exec sp_logEndSection @processName

        exec sp_logEndSection @processName

    end try
    begin catch
        exec sp_logAndRethrowError     
    end catch
end
GO
```

# [dbo].[spr_purgeFactTables]

```sql
/****** Object:  StoredProcedure [dbo].[spr_purgeFactTables]    Script Date: 5/17/2022 11:50:23 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER OFF
GO
create procedure [dbo].[spr_purgeFactTables]
as 
begin

    begin try 
        declare @processName varchar(100)
        -- initialize parameters
        set @processName = 'spr_purgeFactTables'
        declare @factTableName  varchar(255) 
        declare @datePart       varchar(255)   
        declare @periodLen      varchar(255) 
        declare @dateKeyCol     varchar(255) 
        declare @strSQLStmt     nvarchar(max)
        declare @delRowsCount   int
        declare @errorMsg       varchar(255)
        set @errorMsg = '' 

        set @factTableName = (select min(purgeFactTable) from MetaUtilDBPurge)      
        while 1 = 1 
        begin
            exec sp_logWrite @action = 'START', @logLevel = 'INFO',  @processName = @processName, @tblName = @factTableName, @logText = 'Starting of table purging'
            ----------------------------------------------------------------------
            -- getting param values for the given factTable
            ----------------------------------------------------------------------
            select @datePart   = datePart,
                   @dateKeyCol = dateKeyColumn,
                   @periodLen  = periodLen
              from MetaUtilDBPurge
             where purgeFactTable = @factTableName 
            -- table purging
            set @strSQLStmt = 'delete from ' + @factTableName 
            set @strSQLStmt = @strSQLStmt + ' where datediff(' + @datePart + ', convert(datetime, convert(char(8), ' + @dateKeyCol + ')), getdate()) > @pl; set @rc=@@rowcount'
            set @delRowsCount = 0 
            begin try 
                exec sp_executeSQL @strSQLStmt,
                                   N'@pl int, @rc int OUTPUT',
                                   @pl = @periodLen,
                                   @rc = @delRowsCount OUTPUT
                exec sp_logWrite @action = 'STOP', @logLevel = 'INFO', @processName = @processName, @logText = 'Successful Run', @tblName = @factTableName, @deleteCount = @delRowsCount    
            end try
            begin catch
                set @errorMsg = 'Error: the set of parameters: purgeFactTable = '''+ @factTableName +''', ' + 
                                                               'datePart = '''+ @datePart +''', ' + 
                                                               'dateKeyColumn = '''+ @dateKeyCol +''', ' +  
                                                               'periodLength = '''+ @periodLen +''' ' +  
                                 'are incorrect. The purge has not been performed!'
                exec sp_logWrite @logLevel = 'FATAL', @processName = @processName, @logText = @errorMsg, @tblName = @factTableName
            end catch

            
            -- getting next table 
            set @factTableName = (select min(purgeFactTable) from MetaUtilDBPurge where purgeFactTable > @factTableName)
            if @factTableName is null break 
        end 
        /*--------------------------------------------------------------------------------------------*/
        
        if @errorMsg <> '' 
            raiserror (N'Not all purge operation were performed. See logTable for more details.', 16, 1)
        
    end try
    begin catch
        exec sp_logAndRethrowError     
    end catch
    
end
GO
```

# [dbo].[spr_sc_dq_ColumnList]

```sql
/****** Object:  StoredProcedure [dbo].[spr_sc_dq_ColumnList]    Script Date: 5/17/2022 11:50:23 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER OFF
GO
CREATE PROCEDURE [dbo].[spr_sc_dq_ColumnList] 
   @database_name	nvarchar(128),
   @schema_name		nvarchar(256),
   @table_name		nvarchar(256) 
AS
BEGIN
SET NOCOUNT ON

CREATE TABLE #tmp_TableColumns(
	column_name sysname,
	data_type nvarchar(128),
	ordinal_position bigint,
	max_length bigint,
	row_count bigint)

CREATE TABLE #tmp_ColumnProfile(
	TableName varchar(256),
	ColumnName varchar(256),
	OrdinalPosition bigint,
	DataType varchar (60),
	MaxLength bigint,
	MinValue nvarchar(max),
	MaxValue nvarchar(max),
	DistinctValues bigint, 
	NullCount bigint,
	RowsCount bigint,
	AvgLengthOrValue bigint,
	MaxCharLength bigint)

DECLARE 
	@ColName varchar(256), 
	@OrdinalPosition bigint,
	@DataType varchar(60), 
	@MaxLength bigint,
	@RowsCount bigint,
	@MaxLengthTmp bigint,
	@SQLStr varchar(4000)

SET @SQLStr =	'SELECT column_name, data_type, ordinal_position, ISNULL(character_maximum_length,0) + ISNULL(numeric_precision,0) + ISNULL(datetime_precision,0) AS max_length, '+
				'(SELECT COUNT(*) FROM ' + @database_name + '.' + @schema_name + '.[' + @table_name + '])AS row_count ' +
				'FROM ' + @database_name + '.INFORMATION_SCHEMA.COLUMNS ' +
				'WHERE table_schema = ''' + @schema_name + ''' AND table_name =''' + @table_name +''''

INSERT INTO #tmp_TableColumns EXECUTE(@SQLStr)

DECLARE ColCursor CURSOR FOR 
	SELECT * FROM #tmp_TableColumns
	ORDER BY Ordinal_Position

OPEN ColCursor

FETCH NEXT FROM ColCursor 
INTO   @ColName, @DataType, @OrdinalPosition, @MaxLengthTmp, @RowsCount

WHILE @@FETCH_STATUS = 0
BEGIN
	SET @MaxLength = @MaxLengthTmp
	IF @DataType = 'bit' SET @MaxLength = 1
	IF @DataType = 'uniqueidentifier' SET @MaxLength = 16
	IF @DataType = 'datetime' SET @MaxLength = 8
	IF @DataType = 'smalldatetime' SET @MaxLength = 4
	IF @DataType = 'xml' SET @MaxLength = NULL

--	PRINT '>>> Getting Column Name: ''' + @ColName + ''', MaxLength: ' + CAST(ISNULL(@MaxLength,0) AS varchar(10))
-- Create a SQL string to insert column info depending on the data type of the given column
	IF @DataType IN ('int', 'bigint', 'smallint', 'tinyint', 
		'money', 'smallmoney', 'decimal', 'numeric',
		'float', 'real')
	BEGIN  -- handle most numeric data types
		SELECT @SQLStr = 'INSERT INTO #tmp_ColumnProfile SELECT ''' + @table_name +
			''' TableName, ''' + @ColName + ''' ColName, ' + 
			CAST (@OrdinalPosition AS varchar(12)) + ' OrdPos, ''' + @DataType + '(' +
			CAST(ISNULL(  @MaxLength  ,0) AS varchar(12)) + ')' +
			''' DataType, ' + CAST(ISNULL(@MaxLength, 0) AS varchar(12)) + 
			' Lngth, MIN([' + @ColName + ']) MIN, MAX([' + 
			@ColName + ']) MAX, COUNT(DISTINCT [' + @ColName + ']) DistinctVals, ' + 
			'SUM(CASE WHEN [' + @ColName + '] IS NULL THEN 1 ELSE 0 END) NullCount, ' + 
			LTRIM(@RowsCount) +' AS RowsCount, ' +	
			' AVG( [' + @ColName + ']*1.0) AvgValue, ' +  
			' NULL MaxLength' +
			' FROM ' + @database_name + '.' + @schema_name + '.[' + @table_name +']'
	END
	ELSE
	IF @DataType IN ('char', 'varchar', 'nvarchar', 'nchar') -- handle character string data types
	BEGIN 
		SELECT @SQLStr = 'INSERT INTO #tmp_ColumnProfile SELECT ''' + 
			@table_name + ''' TableName, ''' + @ColName + 
			''' ColName, ' + CAST (@OrdinalPosition AS varchar(12)) + 
			' OrdPos, ''' + @DataType + '(' + 
			CAST(ISNULL(  @MaxLength  ,0) AS varchar(12)) + ')' +
			''' DataType, ' + CAST(ISNULL(@MaxLength, 0) AS varchar(12)) + 
			' Lngth, MIN([' + @ColName + ']) MIN, MAX([' + 
			@ColName + ']) MAX, COUNT(DISTINCT [' + @ColName + ']) DistinctVals, ' + 
			'SUM(CASE WHEN [' + @ColName + 
			'] IS NULL THEN 1 ELSE 0 END) NullCount, ' + 
			LTRIM(@RowsCount) +' AS RowsCount, ' +
			' AVG(len( [' + @ColName + ']))  AvgValue, ' +
			' MAX(len( [' + @ColName + ']))  MaxLength ' +  
			' FROM ' + @database_name + '.' + @schema_name + '.[' + @table_name +']'
	END
	ELSE
	IF @DataType IN ('datetime', 'smalldatetime', 'timestamp') -- handle date data types
	BEGIN  
		SELECT @SQLStr = 'INSERT INTO #tmp_ColumnProfile SELECT ''' + @table_name +
			''' TableName, ''' + @ColName + ''' ColName, ' + 
			CAST (@OrdinalPosition AS varchar(12)) + ' OrdPos, ''' + @DataType + '(' +
			CAST(ISNULL(  @MaxLength  ,0) AS varchar(12)) + ')' +
			''' DataType, ' + CAST(ISNULL(@MaxLength, 0) AS varchar(12)) + 
			' Lngth, MIN([' + @ColName + ']) MIN, MAX([' + 
			@ColName + ']) MAX, COUNT(DISTINCT [' + @ColName + ']) DistinctVals, ' + 
			'SUM(CASE WHEN [' + @ColName + '] IS NULL THEN 1 ELSE 0 END) NullCount, ' + 
			LTRIM(@RowsCount) +' AS RowsCount, ' +
			' NULL AvgLength, ' +
			--      ' CAST(Avg(CAST( ' + @ColName + ' AS int)) AS datetime) AvgValue, ' +  
			' NULL MaxLength' +
			' FROM ' + @database_name + '.' + @schema_name + '.[' + @table_name +']'
	END   
	ELSE
	IF @DataType IN ('xml', 'uniqueidentifier', 'binary', 'varbinary', 'image', 'bit', 'text', 'ntext') -- Handle XML and other odd data types
	BEGIN 
		SELECT @SQLStr = 'INSERT INTO #tmp_ColumnProfile SELECT ''' + @table_name +
			''' TableName, ''' + @ColName + ''' ColName, ' + 
			CAST (@OrdinalPosition AS varchar(12)) + ' OrdPos, ''' + @DataType + 
			''' DataType, ' + CAST(ISNULL(@MaxLength, 0) AS varchar(12)) + 
			' Lngth, ''n/a'' MIN, ''n/a'' MAX, 0 DistinctVals, ' + 
			'SUM(CASE WHEN [' + @ColName + '] IS NULL THEN 1 ELSE 0 END) NullCount, ' + 
			LTRIM(@RowsCount) +' AS RowsCount, ' +
			' NULL AvgValue, ' +
			' NULL MaxValue ' +
			' FROM ' + @database_name + '.' + @schema_name + '.[' + @table_name +']'
	END
	ELSE
	BEGIN -- Unidentified data types
		PRINT 'Need a section to handle this data type: ' + @DataType
	END
	EXEC(@SQLStr)
	FETCH NEXT FROM ColCursor 
	INTO @ColName, @DataType, @OrdinalPosition, @MaxLengthTmp, @RowsCount
END -- of cursor column processing
-- Write out the final results for the RS Dataset
SELECT 
	*,
	NullCount / (1.0 * NULLIF(RowsCount,0)) AS PctNull
FROM #tmp_ColumnProfile
ORDER BY OrdinalPosition
-- Clean house
CLOSE ColCursor
DEALLOCATE ColCursor
DROP TABLE #tmp_ColumnProfile
DROP TABLE #tmp_TableColumns
END
GO
```

# [dbo].[spr_sc_dq_TableList]

```sql
/****** Object:  StoredProcedure [dbo].[spr_sc_dq_TableList]    Script Date: 5/17/2022 11:50:23 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER OFF
GO
CREATE PROCEDURE [dbo].[spr_sc_dq_TableList] 
	@database_name nvarchar(128)
AS
BEGIN
SET NOCOUNT ON

CREATE TABLE #tmp_TableProfile
   (id	int,
	full_name   sysname, 
    schema_name sysname,
    table_name  sysname,
    col_count   int)

CREATE TABLE #tmp_SpaceUsed
	(id int,
	reserved_page_count bigint,
	used_page_count bigint,
	pages bigint,
	row_count bigint)


DECLARE @SQLStr nvarchar(4000)

SET @SQLStr =	'SELECT object_id('''+ @database_name + '.''+t.table_schema+''.''+t.table_name), t.table_schema+''.''+t.table_name, t.table_schema, t.table_name, '+
				'COUNT(c.column_name) AS col_count FROM '+ @database_name + 
				'.INFORMATION_SCHEMA.TABLES t INNER JOIN '+ @database_name + 
				'.INFORMATION_SCHEMA.COLUMNS c ON t.table_schema = c.table_schema AND t.table_name = c.table_name ' +
				'WHERE t.table_type = ''BASE TABLE'' GROUP BY t.table_schema, t.table_name '
INSERT INTO #tmp_TableProfile(id,full_name,schema_name,table_name,col_count) EXECUTE(@SQLStr)

SET @SQLStr = 'SELECT 
			object_id,
			--OBJECT_NAME(object_id) table_name,
			SUM(reserved_page_count) + ISNULL(MIN(xml_reserved_page_count),0) AS reserved_page_count,
			SUM(used_page_count) + ISNULL(MIN(xml_used_page_count),0) AS used_page_count,
			SUM(
				CASE
					WHEN (index_id < 2) THEN (in_row_data_page_count + lob_used_page_count + row_overflow_used_page_count)
					ELSE lob_used_page_count + row_overflow_used_page_count
				END
				) AS pages,
			SUM (
				CASE
					WHEN (index_id < 2) THEN row_count
					ELSE 0
				END
				) AS sp_row_count
		FROM ' + @database_name + '.sys.dm_db_partition_stats s LEFT JOIN  
			--
			-- Check if table has XML Indexes or Fulltext Indexes which use internal tables tied to this table
			--
			(SELECT it.parent_id, 
					SUM(resers XML Indexes or Fulltext Indexes which use internal tables tied to this table
			--
			(SELECT it.parent_id, 
					SUM(reserved_page_count) xml_reserved_page_count,
					SUM(used_page_count) xml_used_page_count
				FROM ' + @database_name + '.sys.dm_db_partition_stats p, ' + @database_name + '.sys.internal_tables it
				WHERE it.internal_type IN (202,204) AND p.object_id = it.object_id GROUP BY it.parent_id) x ON x.parent_id = s.object_id 
		WHERE s.object_id IN (SELECT object_id('''+ @database_name + '.''+table_schema+''.''+table_name) FROM '+ @database_name + '.INFORMATION_SCHEMA.TABLES WHERE table_type = ''BASE TABLE'')
		GROUP BY s.object_id' 

INSERT INTO #tmp_SpaceUsed EXECUTE(@SQLStr)

SELECT 
	t.*,
	s.row_count, 
	LTRIM (reserved_page_count * 8) + ' KB' AS reserved_space,
	LTRIM (pages * 8) + ' KB' AS data_space,
	LTRIM ((CASE WHEN used_page_count > pages THEN (used_page_count - pages) ELSE 0 END) * 8) + ' KB' AS index_space,
	LTRIM ((CASE WHEN reserved_page_count > used_page_count THEN (reserved_page_count - used_page_count) ELSE 0 END) * 8) + ' KB' AS unused_space
FROM #tmp_SpaceUsed s INNER JOIN #tmp_TableProfile t ON t.id = s.id
ORDER BY t.table_name

END
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[spr_ScorecardProcessingPriceIndex]    Script Date: 5/17/2022 11:50:23 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE procedure [dbo].[spr_ScorecardProcessingPriceIndex] as
begin


/*
exec spr_ScorecardProcessingPriceIndex

select * from FactOrders

select top 1 * from orders
select top 1 * from FactOrders
select top 1 * from DimProduct
select top 1 * from DimCustomer
select top 1 * from DimRQM
select top 1 * from DimDate
select top 1 * from DimCurrency


select * from logTable where processName = 'spr_ScorecardProcessingPriceIndex'
truncate table logTable
*/

/*
History
==================
06/10/2008	aminkin		Initial creation
*/

/*
Goals
1. Generate PPB lookup for CatalogNuber
2. Generate PPB lookup for productvelocity
3. Generate PPB lookup for prodgroup
4. Generate PPB lookup for prodgroupdv
5. Generate PPB lookup for ProductLineBucket
6. Generate PI lookup for CatalogNumber
7. Generate PI lookup for prodgroup
8. Generate PI lookup for Prodline
*/

--1.	 Add additional ETL
--a.	Add five columns to DImProduct for calculations: . available for each product.  
--b.	Apply Median_DV from PPBRunResults  where possible navigating tree as outlined before
--c.	Go to transactions to find average mediandv where value not found in PPBRun Results
/*
SL	Friendly Name				PGL DimProductDV	DimproductKey		PBRunResultsKey
===============================================================================================
1	Productlinebucket			8	ProdLine_DV		ProductLineBucket	ProductLineBucket
2	new_productgroupbucket		7	ProdGrpVel_DV	prodgroupdv			ProductGroupBucket
3	prodgroup					6	ProdGrp_DV		prodgroup			ProductGroup
4	new_prodvelocity			5	ProdVel_DV		productvelocity		ProductVelocity
5	productid (catalognumber)	4	CatNum_DV		Catalognumber		productkey
*/

--alter table dimproduct add CatNum_DV numeric (16,9)
--alter table dimproduct add ProdGrp_DV numeric (16,9)
--alter table dimproduct add ProdGrpVel_DV numeric (16,9)
--alter table dimproduct add ProdLine_DV numeric (16,9)
--alter table dimproduct add ProdVel_DV numeric (16,9)
--select avg(mediandv) from ppbrunresults
--0.5119158449
/*
For scorecard:
Product Line Search only peergrouplevel =8 in PPBRunResults looking for a match to get mediandv.  
Product Group Search peergrouplevel 6-8 (one query per level, staring with 6) in PPBRunResults until you find a match to get mediandv
Catalog Number Search peergrouplevel 3-8 (one query per level, staring with 3) in PPBRunResults until you find a match to get mediandv
*/

begin try


-- logging parameters
declare @processName varchar(100)
declare @tableName varchar(50)
declare @finalTableRowCount int
declare @logID int
declare @updateCount int
declare @insertCount int
-- initialize parameters
set @processName = 'spr_ScorecardProcessingPriceIndex'
set @tableName = 'FactOrders'


exec sp_logBeginSection @processName
exec sp_logWrite @action = 'START', 
	@logLevel = 'INFO', 
	@processName = @processName, 
	@tblName = @tableName,
	@logText = @processName

				declare @fallbackdv numeric(16,9)
				set @fallbackdv =
				(select avg(((goldbookprice * (1-peerprice))- salesdollars)/salesdollars)
				from factorders f
				where f.quantity > 0 and f.salesdollars > 0 and f.spakey =1)
				--@fallbackdv

        /*--------------------------------------------------------------------------------------------*/
        /*------1. Generate PPB lookup for CatalogNuber-----------------------------------------------*/
        /*--------------------------------------------------------------------------------------------*/

			exec sp_logBeginSection 'Generate PPB lookup for CatalogNuber'
			exec sp_logWrite @action = 'START', 
				@logLevel = 'INFO', 
				@processName = @processName, 
				@tblName = @tableName,
				@logText = @processName


				Update dimproduct set CatNum_DV =P.mediandv
				from dimproduct R
				join 
				ppbrunresults P
				ON	P.ProductLineBucket		= R.ProductLineBucket		
				AND	P.ProductGroupBucket	= R.prodgroupdv		
				AND	P.ProductGroup			= R.prodgroup		
				AND	P.ProductVelocity		= R.productvelocity	
				AND	P.productkey			= R.Catalognumber	
				WHERE CatNum_DV IS NULL
				AND peergrouplevel = 4

				Update dimproduct set CatNum_DV =P.mediandv
				from dimproduct R
				join 
				ppbrunresults P
				ON	P.ProductLineBucket		= R.ProductLineBucket		
				AND	P.ProductGroupBucket	= R.prodgroupdv		
				AND	P.ProductGroup			= R.prodgroup		
				AND	P.ProductVelocity		= R.productvelocity	
				--AND	P.productkey			= R.Catalognumber	
				WHERE CatNum_DV IS NULL
				AND peergrouplevel = 5

				Update dimproduct set CatNum_DV =P.mediandv
				from dimproduct R
				join 
				ppbrunresults P
				ON	P.ProductLineBucket		= R.ProductLineBucket		
				AND	P.ProductGroupBucket	= R.prodgroupdv		
				AND	P.ProductGroup			= R.prodgroup		
				--AND	P.ProductVelocity		= R.productvelocity	
				--AND	P.productkey			= R.Catalognumber	
				WHERE CatNum_DV IS NULL
				AND peergrouplevel = 6

				Update dimproduct set CatNum_DV =P.mediandv
				from dimproduct R
				join 
				ppbrunresults P
				ON	P.ProductLineBucket		= R.ProductLineBucket		
				AND	P.ProductGroupBucket	= R.prodgroupdv		
				--AND	P.ProductGroup			= R.prodgroup		
				--AND	P.ProductVelocity		= R.productvelocity	
				--AND	P.productkey			= R.Catalognumber	
				WHERE CatNum_DV IS NULL
				AND peergrouplevel = 7

				Update dimproduct set CatNum_DV =P.mediandv
				from dimproduct R
				join 
				ppbrunresults P
				ON	P.ProductLineBucket		= R.ProductLineBucket		
				WHERE CatNum_DV IS NULL
				AND peergrouplevel = 8

				IF  EXISTS (SELECT * FROM sys.objects WHERE object_id 
				= OBJECT_ID(N'[dbo].[prodkeydv]'))
				DROP table prodkeydv

				select f.productkey, avg(((goldbookprice * (1-peerprice))- salesdollars)/salesdollars) peerprice
				into prodkeydv
				from dimproduct p
				join factorders f
				on f.productkey = p.productkey
				where p.CatNum_DV is null
				and f.quantity > 0 and f.salesdollars > 0 and spakey =1  and goldbookprice > 0
				group by f.productkey

				update dimproduct set CatNum_DV = f.peerprice
				from dimproduct p
				join prodkeydv f
				on f.productkey = p.productkey
				where CatNum_DV is null

				update dimproduct 
				set CatNum_DV  = @fallbackdv
				where CatNum_DV is null or CatNum_DV = 0

			set @insertcount= @@rowcount
			exec sp_logWrite @action = 'STOP', 
				@logLevel = 'INFO', 
				@processName = @processName, 
				@logText = 'Successful Run', 
				@tblName = @tableName,
				@updateCount = @updateCount
			exec sp_logEndSection 'Generate PPB lookup for CatalogNuber'

        /*--------------------------------------------------------------------------------------------*/
        /*------2. Generate PPB lookup for productvelocity-----------------------------------------------*/
        /*--------------------------------------------------------------------------------------------*/

			exec sp_logBeginSection 'Generate PPB lookup for productvelocity'
			exec sp_logWrite @action = 'START', 
				@logLevel = 'INFO', 
				@processName = @processName, 
				@tblName = @tableName,
				@logText = @processName

				/*
				============================================================================
				PRODUCT VELOCITY
				============================================================================
				*/

				Update dimproduct set ProdVel_DV =P.mediandv
				from dimproduct R
				join 
				ppbrunresults P
				ON	P.ProductLineBucket		= R.ProductLineBucket		
				AND	P.ProductGroupBucket	= R.prodgroupdv		
				AND	P.ProductGroup			= R.prodgroup		
				AND	P.ProductVelocity		= R.productvelocity	
				WHERE ProdVel_DV IS NULL
				AND peergrouplevel = 5

				Update dimproduct set ProdVel_DV =P.mediandv
				from dimproduct R
				join 
				ppbrunresults P
				ON	P.ProductLineBucket		= R.ProductLineBucket		
				AND	P.ProductGroupBucket	= R.prodgroupdv		
				AND	P.ProductGroup			= R.prodgroup		
				WHERE ProdVel_DV IS NULL
				AND peergrouplevel = 6

				Update dimproduct set ProdVel_DV =P.mediandv
				from dimproduct R
				join 
				ppbrunresults P
				ON	P.ProductLineBucket		= R.ProductLineBucket		
				AND	P.ProductGroupBucket	= R.prodgroupdv			
				WHERE ProdVel_DV IS NULL
				AND peergrouplevel = 7

				Update dimproduct set ProdVel_DV =P.mediandv
				from dimproduct R
				join 
				ppbrunresults P
				ON	P.ProductLineBucket		= R.ProductLineBucket			
				WHERE ProdVel_DV IS NULL
				AND peergrouplevel = 8


				IF  EXISTS (SELECT * FROM sys.objects WHERE object_id 
				= OBJECT_ID(N'[dbo].[prodvelokeydv]'))
				DROP table prodvelokeydv

				select p.productvelocity, avg(((goldbookprice * (1-peerprice))- salesdollars)/salesdollars) peerprice
				into prodvelokeydv
				from dimproduct p
				join factorders f
				on f.productkey = p.productkey
				where p.ProdVel_DV is null
				and f.quantity > 0 and f.salesdollars > 0 and spakey =1 and goldbookprice > 0
				group by p.productvelocity

				update dimproduct set ProdVel_DV = f.peerprice
				from dimproduct p
				join prodvelokeydv f
				on f.productvelocity = p.productvelocity
				where ProdVel_DV is null or ProdVel_DV = 0

				update dimproduct 
				set ProdVel_DV  = @fallbackdv
				where ProdVel_DV is null or ProdVel_DV = 0

			set @insertcount= @@rowcount
			exec sp_logWrite @action = 'STOP', 
				@logLevel = 'INFO', 
				@processName = @processName, 
				@logText = 'Successful Run', 
				@tblName = @tableName,
				@updateCount = @updateCount
			exec sp_logEndSection 'Generate PPB lookup for productvelocity'

        /*--------------------------------------------------------------------------------------------*/
        /*------3. Generate PPB lookup for prodgroup-----------------------------------------------*/
        /*--------------------------------------------------------------------------------------------*/

			exec sp_logBeginSection 'Generate PPB lookup for prodgroup'
			exec sp_logWrite @action = 'START', 
				@logLevel = 'INFO', 
				@processName = @processName, 
				@tblName = @tableName,
				@logText = @processName

				/*
				============================================================================
				PRODUCT GROUP ProdGrp_DV, ProdGrpVel_DV, ProdLine_DV
				============================================================================
				*/

				Update dimproduct set ProdGrp_DV =P.mediandv
				from dimproduct R
				join 
				ppbrunresults P
				ON	P.ProductLineBucket		= R.ProductLineBucket		
				AND	P.ProductGroupBucket	= R.prodgroupdv		
				AND	P.ProductGroup			= R.prodgroup			
				WHERE ProdGrp_DV IS NULL
				AND peergrouplevel = 6

				Update dimproduct set ProdGrp_DV =P.mediandv
				from dimproduct R
				join 
				ppbrunresults P
				ON	P.ProductLineBucket		= R.ProductLineBucket		
				AND	P.ProductGroupBucket	= R.prodgroupdv		
				WHERE ProdGrp_DV IS NULL
				AND peergrouplevel = 7

				Update dimproduct set ProdGrp_DV =P.mediandv
				from dimproduct R
				join 
				ppbrunresults P
				ON	P.ProductLineBucket		= R.ProductLineBucket		
				WHERE ProdGrp_DV IS NULL
				AND peergrouplevel = 8

				IF  EXISTS (SELECT * FROM sys.objects WHERE object_id 
				= OBJECT_ID(N'[dbo].[prodgrpkeydv]'))
				drop table prodgrpkeydv

				select p.prodgroup, avg(((goldbookprice * (1-peerprice))- salesdollars)/salesdollars) peerprice
				into prodgrpkeydv
				from dimproduct p
				join factorders f
				on f.productkey = p.productkey
				where p.ProdGrp_DV is null
				and f.quantity > 0 and f.salesdollars > 0 and spakey =1  and goldbookprice > 0
				group by p.prodgroup

				update dimproduct set ProdGrp_DV = f.peerprice
				from dimproduct p
				join prodgrpkeydv f
				on f.prodgroup = p.prodgroup
				where ProdGrp_DV is null or ProdGrp_DV = 0

				update dimproduct 
				set ProdGrp_DV  = @fallbackdv
				where ProdGrp_DV is null or ProdGrp_DV = 0

			set @insertcount= @@rowcount
			exec sp_logWrite @action = 'STOP', 
				@logLevel = 'INFO', 
				@processName = @processName, 
				@logText = 'Successful Run', 
				@tblName = @tableName,
				@updateCount = @updateCount
			exec sp_logEndSection 'Generate PPB lookup for prodgroup'

        /*--------------------------------------------------------------------------------------------*/
        /*------4. Generate PPB lookup for prodgroupdv-----------------------------------------------*/
        /*--------------------------------------------------------------------------------------------*/

			exec sp_logBeginSection 'Generate PPB lookup for prodgroupdv'
			exec sp_logWrite @action = 'START', 
				@logLevel = 'INFO', 
				@processName = @processName, 
				@tblName = @tableName,
				@logText = @processName

				/*
				============================================================================
				PRODUCT GROUP BUCKET ProdGrpVel_DV, ProdLine_DV
				============================================================================
				*/

				Update dimproduct set ProdGrpVel_DV =P.mediandv
				from dimproduct R
				join 
				ppbrunresults P
				ON	P.ProductLineBucket		= R.ProductLineBucket		
				AND	P.ProductGroupBucket	= R.prodgroupdv		
				WHERE ProdGrpVel_DV IS NULL
				AND peergrouplevel = 7

				Update dimproduct set ProdGrpVel_DV =P.mediandv
				from dimproduct R
				join 
				ppbrunresults P
				ON	P.ProductLineBucket		= R.ProductLineBucket		
				WHERE ProdGrpVel_DV IS NULL
				AND peergrouplevel = 8

				IF  EXISTS (SELECT * FROM sys.objects WHERE object_id 
				= OBJECT_ID(N'[dbo].[prodgrplvlkeydv]'))
				drop table prodgrplvlkeydv

				select p.prodgroupdv, avg(((goldbookprice * (1-peerprice))- salesdollars)/salesdollars) peerprice
				into prodgrplvlkeydv
				from dimproduct p
				join factorders f
				on f.productkey = p.productkey
				where p.ProdGrpVel_DV is null
				and f.quantity > 0 and f.salesdollars > 0 and spakey =1  and goldbookprice > 0
				group by p.prodgroupdv

				update dimproduct set ProdGrpVel_DV = f.peerprice
				from dimproduct p
				join prodgrplvlkeydv f
				on f.prodgroupdv = p.prodgroupdv
				where ProdGrpVel_DV is null or ProdGrpVel_DV = 0

				update dimproduct 
				set ProdGrpVel_DV  = @fallbackdv
				where ProdGrpVel_DV is null or ProdGrpVel_DV = 0

			set @insertcount= @@rowcount
			exec sp_logWrite @action = 'STOP', 
				@logLevel = 'INFO', 
				@processName = @processName, 
				@logText = 'Successful Run', 
				@tblName = @tableName,
				@updateCount = @updateCount
			exec sp_logEndSection 'Generate PPB lookup for prodgroupdv'

        /*--------------------------------------------------------------------------------------------*/
        /*------5. Generate PPB lookup for ProductLineBucket-----------------------------------------------*/
        /*--------------------------------------------------------------------------------------------*/

			exec sp_logBeginSection 'Generate PPB lookup for ProductLineBucket'
			exec sp_logWrite @action = 'START', 
				@logLevel = 'INFO', 
				@processName = @processName, 
				@tblName = @tableName,
				@logText = @processName

				/*
				============================================================================
				PRODUCT LINE ProdLine_DV
				============================================================================
				*/

				Update dimproduct set ProdLine_DV =P.mediandv
				from dimproduct R
				join 
				ppbrunresults P
				ON	P.ProductLineBucket		= R.ProductLineBucket			
				WHERE ProdLine_DV IS NULL
				AND peergrouplevel = 8

				IF  EXISTS (SELECT * FROM sys.objects WHERE object_id 
				= OBJECT_ID(N'[dbo].[prodlinekeydv]'))
				drop table prodlinekeydv

				select p.ProductLineBucket, avg(((goldbookprice * (1-peerprice))- salesdollars)/salesdollars) peerprice
				into prodlinekeydv
				from dimproduct p
				join factorders f
				on f.productkey = p.productkey
				where p.ProdLine_DV is null
				and f.quantity > 0 and f.salesdollars > 0 and spakey =1  and goldbookprice > 0
				group by p.ProductLineBucket

				update dimproduct set ProdLine_DV = f.peerprice
				from dimproduct p
				join prodlinekeydv f
				on f.ProductLineBucket = p.ProductLineBucket
				where ProdLine_DV is null or ProdLine_DV = 0
				--sp_who
				update dimproduct 
				set ProdLine_DV  = @fallbackdv
				where ProdLine_DV is null or ProdLine_DV = 0

			set @insertcount= @@rowcount
			exec sp_logWrite @action = 'STOP', 
				@logLevel = 'INFO', 
				@processName = @processName, 
				@logText = 'Successful Run', 
				@tblName = @tableName,
				@updateCount = @updateCount
			exec sp_logEndSection 'Generate PPB lookup for ProductLineBucket'

        /*--------------------------------------------------------------------------------------------*/
        /*------6. Generate Scorecard lookup for CatalogNumber----------------------------------------*/
        /*--------------------------------------------------------------------------------------------*/

			exec sp_logBeginSection 'Generate Scorecard lookup for CatalogNumber'
			exec sp_logWrite @action = 'START', 
				@logLevel = 'INFO', 
				@processName = @processName, 
				@tblName = @tableName,
				@logText = @processName


			IF  EXISTS (SELECT * FROM sys.objects WHERE object_id = OBJECT_ID(N'[dbo].[cn_disc]') AND type in (N'U'))
			drop table cn_disc

				CREATE CLUSTERED INDEX ix_date on cn_disc(txnmonth)

				select 
				convert(int,round(datekey/100,0,1)) txnmonth, 
				s.BusinessUnitId bu,
				s.CustRegCode + ' - '+ s.CustRegDesc region,
				s.salesofficecode + ' - '+s.SalesOfficedesc SalesOffice,  
				p.catalognumber, 
				sum(f.goldbookprice) goldbookprice , 
				sum(f.salesdollars) salesdollars,
				count(*) txncount
				into cn_disc
				from factorders f
				join dimproduct p
				on 
				p.productkey = f.productkey
				join dimcustomer s
				on 
				f.customerkey = s.customerkey
				where salesdollars > 0  and goldbookprice > 0
				group by 
				convert(int,round(datekey/100,0,1)), 
				s.BusinessUnitId,
				s.CustRegCode + ' - '+ s.CustRegDesc,
				s.salesofficecode + ' - '+s.salesofficedesc, 
				p.catalognumber

			set @insertcount= @@rowcount

			CREATE INDEX ix_bu			on cn_disc(bu)
			CREATE INDEX ix_product		on cn_disc(catalognumber)
			CREATE INDEX ix_region		on cn_disc(Region)
			CREATE INDEX ix_salesoffice	on cn_disc(SalesOffice)

			exec sp_logWrite @action = 'STOP', 
				@logLevel = 'INFO', 
				@processName = @processName, 
				@logText = 'Successful Run', 
				@tblName = @tableName,
				@updateCount = @updateCount
			exec sp_logEndSection 'Generate Scorecard lookup for CatalogNumber'


        /*--------------------------------------------------------------------------------------------*/
        /*------7. Generate Scorecard lookup for prodgroup--------------------------------------------*/
        /*--------------------------------------------------------------------------------------------*/

			exec sp_logBeginSection 'Generate Scorecard lookup for prodgroup'
			exec sp_logWrite @action = 'START', 
				@logLevel = 'INFO', 
				@processName = @processName, 
				@tblName = @tableName,
				@logText = @processName

				IF  EXISTS (SELECT * FROM sys.objects WHERE object_id = OBJECT_ID(N'[dbo].[pg_disc]') AND type in (N'U'))
				drop table pg_disc
				
				CREATE CLUSTERED INDEX ix_date on pg_disc(txnmonth)

				select 
				convert(int,round(datekey/100,0,1)) txnmonth, 
				s.BusinessUnitId bu,
				s.CustRegCode + ' - '+ s.CustRegDesc Region,
				s.salesofficecode + ' - '+s.SalesOfficedesc SalesOffice,  
				p.prodgroup, 
				sum(f.goldbookprice) goldbookprice , 
				sum(f.salesdollars) salesdollars,
				count(*) txncount
				into pg_disc
				from factorders f
				join dimproduct p
				on 
				p.productkey = f.productkey
				join dimcustomer s
				on 
				f.customerkey = s.customerkey
				where salesdollars > 0  and goldbookprice > 0
				group by 
				convert(int,round(datekey/100,0,1)), 
				s.BusinessUnitId ,
				s.CustRegCode + ' - '+ s.CustRegDesc ,
				s.salesofficecode + ' - '+s.SalesOfficedesc , 
				p.prodgroup

			set @insertcount= @@rowcount

			CREATE INDEX ix_bu			on pg_disc(bu)
			CREATE INDEX ix_product		on pg_disc(prodgroup)
			CREATE INDEX ix_region		on pg_disc(Region)
			CREATE INDEX ix_salesoffice	on pg_disc(SalesOffice)

			exec sp_logWrite @action = 'STOP', 
				@logLevel = 'INFO', 
				@processName = @processName, 
				@logText = 'Successful Run', 
				@tblName = @tableName,
				@updateCount = @updateCount
			exec sp_logEndSection 'Generate Scorecard lookup for prodgroup'


        /*--------------------------------------------------------------------------------------------*/
        /*------8. Generate Scorecard lookup for Prodline---------------------------------------------*/
        /*--------------------------------------------------------------------------------------------*/

			exec sp_logBeginSection 'Generate Scorecard lookup for Prodline'
			exec sp_logWrite @action = 'START', 
				@logLevel = 'INFO', 
				@processName = @processName, 
				@tblName = @tableName,
				@logText = @processName

				IF  EXISTS (SELECT * FROM sys.objects WHERE object_id = OBJECT_ID(N'[dbo].[pl_disc]') AND type in (N'U'))
				drop table pl_disc

				CREATE CLUSTERED INDEX ix_date on pl_disc(txnmonth)

				select 
				convert(int,round(datekey/100,0,1)) txnmonth, 
				s.BusinessUnitId bu,
				s.CustRegCode + ' - '+ s.CustRegDesc Region,
				s.salesofficecode + ' - '+s.SalesOfficedesc SalesOffice,  
				p.prodline, 
				sum(f.goldbookprice) goldbookprice , 
				sum(f.salesdollars) salesdollars,
				count(*) txncount
				into pl_disc
				from factorders f
				join dimproduct p
				on 
				p.productkey = f.productkey
				join dimcustomer s
				on 
				f.customerkey = s.customerkey
				where salesdollars > 0  and goldbookprice > 0
				group by 
				convert(int,round(datekey/100,0,1)), 
				s.BusinessUnitId,
				s.CustRegCode + ' - '+ s.CustRegDesc,
				s.salesofficecode + ' - '+s.SalesOfficedesc , 
				p.prodline

			set @insertcount= @@rowcount


			CREATE INDEX ix_bu			on pl_disc(bu)
			CREATE INDEX ix_product		on pl_disc(prodgroup)
			CREATE INDEX ix_region		on pl_disc(Region)
			CREATE INDEX ix_salesoffice	on pl_disc(SalesOffice)

			exec sp_logWrite @action = 'STOP', 
				@logLevel = 'INFO', 
				@processName = @processName, 
				@logText = 'Successful Run', 
				@tblName = @tableName,
				@updateCount = @updateCount
			exec sp_logEndSection 'Generate Scorecard lookup for Prodline'

/*----------------------------------END----------------------------*/

			exec sp_logWrite @action = 'STOP', 
				@logLevel = 'INFO', 
				@processName = @processName, 
				@logText = 'Successful Run', 
				@tblName = @tableName
			exec sp_logEndSection @processName

    end try
    begin catch
        exec sp_logAndRethrowError     
    end catch
end
GO
```
